```python
#pytho畫圖5  #加上z
import matplotlib.pyplot as pt
import numpy as np
x=np.arange(0,360)
y=np.sin(x*np.pi/180.0)
z=np.sin(x*np.pi/90.0)
pt.xlim(0,360)
pt.ylim(-1.2,1.2)
pt.plot(x,y)
pt.plot(x,z)
pt.title("Sin & Cos wave")
pt.xlabel("Degree")
pt.ylabel("Value")
pt.show()
```

```python
#k-means群聚clustering演算法
from sklearn import datasets,cluster,metrics
                    # 資料集 # 計算  #驗證偵測
iris=datasets.load_iris()
iris_data=iris.data
#執行k-means群聚演算法
result=cluster.KMeans(n_clusters=3).fit(iris_data)
#result.labels_預測的分群結果
print(result.labels_)
#用silhouette score評估分群結果好壞，越接近1越好
silhouette_avg=metrics.silhouete_score(iris_data,result.labels_)
print(silhouette_avg)
```

```python
from sklearn import datasets,svm
import matplotlib.pyplot as plt
digits=datasets.load_digits()
from sklearn.cross_validation import train_test_split
X_train,X_test,y_train,y_test,images_train,images_test =train_test_split(digits.data,digits.target,digits.images,test_size=0.25,random_state=42)
svc_model = svm.SVC(gamma=0.001,C=100.,kernel='linear')
svc_model.fit(X_train,y_train)
predicted = svc_model.predict(X_test)
images_and_predictions = list(zip(images_test,predicted))
for index,(image,prediction) in enumerate(images_and_predictions[:5]):
    plt.subplot(1,5,index + 1)
    plt.axis('off')
    plt.imshow(image, cmap=plt.cm.binary)
    plt.title('Predicted: ' + str(prediction))
plt.show()
```
